## 2. EDA

### 2.1. Analyse univariée

#### 2.1.1 Variable cible

```python
print(f"Nombre de classes uniques : {df['label'].nunique()}")
print("Valeurs possibles :", df['label'].unique())
```

```python
label_counts = df["label"].value_counts()
label_counts
```

```python
label_mapping = {
    'Benign': 'Benign',

    'PartOfAHorizontalPortScan': 'Scan_Attack',
    'Attack': 'Scan_Attack',

    'Okiru': 'Malware_Okiru',
    'Okiru-Attack': 'Malware_Okiru',

    'DDoS': 'DDoS',

    'C&C': 'C&C',
    'C&C-HeartBeat': 'C&C',
    'C&C-FileDownload': 'C&C',
    'C&C-HeartBeat-FileDownload': 'C&C',
    'C&C-Torii': 'C&C',
    'C&C-Mirai': 'C&C',
    'FileDownload': 'C&C'
}
```

```python
df['label'] = df['label'].replace(label_mapping)
```

```python
label_counts = df["label"].value_counts()
label_counts
```

#### 2.1.2 Variables catégorielles

```python
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    print(f"\n{col}")
    print("Unique values:", df[col].nunique())
    print(df[col].value_counts().head(10))
```

```python
df = df.drop(columns=['Unnamed: 0', 'ts' , 'uid' , 'id.orig_h' , 'id.resp_h' , 'local_orig', 'local_resp' , 'history'])
```

```python
# Colonnes numériques : remplacer '-' par 0 et convertir
for col in ['duration','orig_bytes','resp_bytes']:
    df[col] = df[col].replace('-', 0).astype(float)
```

```python
#transformer en catégorie
df = encode_categories(df)
```

```python
df.info()
```

```python
cat_cols = df.select_dtypes(include="category").columns
for col in cat_cols:
    print(f"\n{col}")
    print("Unique values:", df[col].nunique())
    print(df[col].value_counts().head(10))
```

#### 2.2 Analyse bivariée

```python
for col in ['proto', 'service', 'conn_state']:
    # Créer un crosstab normalisé
    ct = pd.crosstab(df[col], df['label'], normalize='index') * 100
    
    plt.figure(figsize=(12, 6))
    ct.plot(kind='bar', stacked=True, ax=plt.gca())
    plt.xlabel(col)
    plt.ylabel('Pourcentage (%)')
    plt.legend(title='Label', bbox_to_anchor=(1.05, 1))
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# 1. On prend les 5 colonnes de chiffres (sans les ports)
cols = ['duration', 'orig_bytes', 'resp_bytes', 'orig_pkts', 'resp_pkts']

# 2. On affiche la carte de chaleur
plt.figure(figsize=(8, 6))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Est-ce que mes variables se répètent ? (1.0 = Oui)")
plt.show()
```
